## Analysis

### Load Packages and Data

```python
# packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.io as pio
import seaborn as sns
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, Holt
from statsmodels.tsa.stattools import adfuller

# settings
%matplotlib inline
```

```python
# data
capacity = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-05-03/capacity.csv')
wind = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-05-03/wind.csv')
solar = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-05-03/solar.csv')
average_cost = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-05-03/average_cost.csv')
```

### Tidy Data Sets

```python
# aggregate wind to month
wind['date'] = pd.to_datetime(wind['date'])
wind['month'] = wind['date'] - pd.offsets.MonthBegin(1)
wind_agg = wind.groupby(['month']).agg({'wind_mwh': 'mean', 'wind_capacity': 'mean'}).reset_index()

# aggregate solar to month
solar['date'] = pd.to_datetime(solar['date'])
solar['month'] = solar['date'] - pd.offsets.MonthBegin(1)
solar_agg = solar.groupby(['month']).agg({'solar_mwh': 'mean', 'solar_capacity': 'mean'})

# join together
ws = wind_agg.merge(solar_agg, how = 'inner', on = 'month')

# pivot capacity longer
ws_cap = ws[['month', 'wind_capacity', 'solar_capacity']].melt(id_vars = ['month'], value_vars = ['solar_capacity', 'wind_capacity'])
ws_cap['variable'] = ws_cap['variable'].replace({'solar_capacity': 'solar', 'wind_capacity': 'wind'})

# pivot mwh longer
ws_mwh = ws[['month', 'wind_mwh', 'solar_mwh']].melt(id_vars = ['month'], value_vars = ['solar_mwh', 'wind_mwh'])
ws_mwh['variable'] = ws_cap['variable'].replace({'solar_mwh': 'solar', 'wind_mwh': 'wind'})
```

```python
sns.relplot(
    data = ws_mwh, x = "month", y = "value", 
    hue = "variable", kind = "line",
    height = 5, aspect = 2, palette = "gist_earth"
).set(
    title = "Comparison of Wind and Solar Costs (Dollar / MWh) Month over Month",
    xlabel = "Date",
    ylabel = "Dollars / MWh"
);
```

## Perform Exponential Smoothing

Use Exponential Smoothing if not Stationary. Use ARIMA if stationary.

```python
# separate out time series
ws_wind_cap = pd.DataFrame(ws, columns = ['month', 'wind_capacity']).set_index('month')
ws_wind_cap.index = pd.to_datetime(ws_wind_cap.index)

ws_solar_cap = pd.DataFrame(ws, columns = ['month', 'solar_capacity']).set_index('month')
ws_solar_cap.index = pd.to_datetime(ws_solar_cap.index)

ws_wind_mwh = pd.DataFrame(ws, columns = ['month', 'wind_mwh']).set_index('month')
ws_wind_mwh.index = pd.to_datetime(ws_wind_mwh.index)

ws_solar_mwh = pd.DataFrame(ws, columns = ['month', 'solar_mwh']).set_index('month')
ws_solar_mwh.index = pd.to_datetime(ws_solar_mwh.index)

# stationary test for Wind Capacity
print(" > Is the Wind Capacity stationary ?")
dftest = adfuller(ws_wind_cap.wind_capacity, autolag = 'AIC')
print("Test statistic = {:.3f}".format(dftest[0]))
print("P-value = {:.3f}".format(dftest[1]))
print("Critical values :")
for k, v in dftest[4].items():
    print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
    
# stationary test for Solar Capacity
print(" > Is the Wind Capacity stationary ?")
dftest = adfuller(ws_solar_cap.solar_capacity, autolag = 'AIC')
print("Test statistic = {:.3f}".format(dftest[0]))
print("P-value = {:.3f}".format(dftest[1]))
print("Critical values :")
for k, v in dftest[4].items():
    print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
    
# stationary test for Wind MWh
print(" > Is the Wind MWh stationary ?")
dftest = adfuller(ws_wind_mwh.wind_mwh, autolag = 'AIC')
print("Test statistic = {:.3f}".format(dftest[0]))
print("P-value = {:.3f}".format(dftest[1]))
print("Critical values :")
for k, v in dftest[4].items():
    print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
    
# stationary test for Solar MWh
print(" > Is the Wind MWh stationary ?")
dftest = adfuller(ws_solar_mwh.solar_mwh, autolag = 'AIC')
print("Test statistic = {:.3f}".format(dftest[0]))
print("P-value = {:.3f}".format(dftest[1]))
print("Critical values :")
for k, v in dftest[4].items():
    print("\t{}: {} - The data is {} stationary with {}% confidence".format(k, v, "not" if v<dftest[0] else "", 100-int(k[:-1])))
```

__Conclusion:__ Capacity is stationary while Cost is not.

```python
# define pred DatetimeIndex
pred_index = pd.date_range(ws_wind_mwh.index[-1] + pd.DateOffset(months = 1), periods = 12, freq = 'MS')
```

```python
# smooth out wind wmh
## train an exponential smoothing model on data
wind_model1 = SimpleExpSmoothing(np.asarray(ws_wind_mwh['wind_mwh']))
fit1 = wind_model1.fit()

### retrieve smoothed and predicted values out for 12 months
ws_wind_wmh_smooth_fit = fit1.fittedvalues
ws_wind_wmh_smooth_pred = fit1.forecast(12)

### redefine data frame

#### fitted values
ws_wind_wmh_df = pd.DataFrame(ws_wind_wmh_smooth_fit, columns = ['wind_mwh_smooth'])
ws_wind_wmh_df.index = ws_wind_mwh.index
ws_wind_mwh2 = ws_wind_wmh_df.merge(ws_wind_mwh, how = 'inner', left_index = True, right_index = True)

#### pred values
ws_wind_wmh_pred_df = pd.DataFrame(ws_wind_wmh_smooth_pred, columns = ['wind_mwh_smooth'])
ws_wind_wmh_pred_df.index = pred_index
ws_wind = pd.concat([ws_wind_mwh2, ws_wind_wmh_pred_df])
```

```python
# smooth out solar wmh
## train an exponential smoothing model on data
solar_model1 = SimpleExpSmoothing(np.asarray(ws_solar_mwh['solar_mwh']))
fit2 = solar_model1.fit()

### retrieve smoothed and predicted values out for 12 months
ws_solar_wmh_smooth_fit = fit2.fittedvalues
ws_solar_wmh_smooth_pred = fit2.forecast(12)

### redefine data frame

#### fitted values
ws_solar_wmh_df = pd.DataFrame(ws_solar_wmh_smooth_fit, columns = ['solar_mwh_smooth'])
ws_solar_wmh_df.index = ws_solar_mwh.index
ws_solar_mwh2 = ws_solar_wmh_df.merge(ws_solar_mwh, how = 'inner', left_index = True, right_index = True)

#### pred values
ws_solar_wmh_pred_df = pd.DataFrame(ws_solar_wmh_smooth_pred, columns = ['solar_mwh_smooth'])
ws_solar_wmh_pred_df.index = pred_index
ws_solar = pd.concat([ws_solar_mwh2, ws_solar_wmh_pred_df])
```

```python
# combine data frames

## clean up wind smooth
ws_wind = ws_wind.rename(columns = {'wind_mwh': 'actual', 'wind_mwh_smooth': 'smooth'})
ws_wind['tense'] = np.where(ws_wind['actual'].isna(), 'Future', 'Past')
ws_wind['category'] = 'Wind'

## clean up solar smooth
ws_solar = ws_solar.rename(columns = {'solar_mwh': 'actual', 'solar_mwh_smooth': 'smooth'})
ws_solar['tense'] = np.where(ws_solar['actual'].isna(), 'Future', 'Past')
ws_solar['category'] = 'Solar'

mwh = pd.concat([ws_wind, ws_solar]).reset_index()
mwh
```
